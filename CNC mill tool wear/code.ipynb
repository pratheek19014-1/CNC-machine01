```python
import kagglehub
import plotly.express as px
import pandas as pd
# Download latest version
path = kagglehub.dataset_download("shasun/tool-wear-detection-in-cnc-mill")
print("Path to dataset files:", path)
```

```python
# store the data in df

df = pd.read_csv("E:/TH_koeln_AIT/Courses/Sem_4/Datasets/Tool_wear_detection/train/experiment_01.csv")
df = df.drop(columns=["Machining_Process"])
print(df.head())
```

```python
# display column names with its index for reference

for idx, col in enumerate(df.columns):
    print(f"{idx}: {col}")
```

```python
# displaying the first column distribution with plotly

# for column in df.columns:
#     print(f"Processing column: {column}")
#     fig = px.histogram(df, x=column, nbins=50, title=f"Distribution of {column}")
#     fig.add_vline(x=df[column].mean(), line_width=4, line_color="green", annotation_text="Mean", annotation_position="top right")
#     fig.add_vline(x=df[column].mean() + 2*df[column].std(), line_width=4, line_color="orange", annotation_text="+2 SD", annotation_position="top right")
#     fig.write_image(column + "_distribution.png")
#     fig.show()

print("Generating distribution plot for first column")
fig = px.histogram(df, x=df.columns[0], nbins=50, title=f"Distribution of {df.columns[0]}")
column = df.columns[0]
fig.add_vline(x=df[column].mean(), line_width=4, line_color="green", annotation_text="Mean", annotation_position="top right")
fig.add_vline(x=df[column].mean() + 2*df[column].std(), line_width=4, line_color="orange", annotation_text="+2 SD", annotation_position="top right")
fig.add_vline(x=df[column].mean() - 2*df[column].std(), line_width=4, line_color="orange", annotation_text="-2 SD", annotation_position="top left")
fig.add_vline(x=df[column].mean() + 3*df[column].std(), line_width=4, line_color="red", annotation_text="+3 SD", annotation_position="top right")
fig.add_vline(x=df[column].mean() - 3*df[column].std(), line_width=4, line_color="red", annotation_text="-3 SD", annotation_position="top left")
fig.show()
```
